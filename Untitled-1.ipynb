```python
import pandas as pd

df = pd.read_excel("Online Retail.xlsx")
df.head()
```

```python
df=df.dropna(subset=['CustomerID'])
df=df[(df['Quantity']>0)&(df['UnitPrice']>0)]
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
df['Sales']=df['Quantity']*df['UnitPrice']
```

```python
import datetime as dt
snapshot_date=df['InvoiceDate'].max()+dt.timedelta(days=1)
rfm=df.groupby('CustomerID').agg({'InvoiceDate':lambda x:(snapshot_date-x.max()).days,
                                  'InvoiceNo':'nunique',
                                  'Sales':'sum'})
rfm.columns=['Recency','Frequency','Monetary']
rfm=rfm.reset_index()
```

```python
import numpy as np
rfm_log=rfm.copy()
rfm_log[['Recency','Frequency','Monetary']]=np.log1p(rfm_log[['Recency','Frequency','Monetary']])
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
rfm_scaled=scaler.fit_transform(rfm_log[['Recency','Frequency','Monetary']])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for i in range(2, 8):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(rfm_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(2, 8), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
cluster_profile = rfm.groupby('Cluster').mean()
print(cluster_profile)
```

```python
def assign_segment(row):
    if row['Monetary'] >= rfm['Monetary'].quantile(0.75):
        return "High Value"
    elif row['Frequency'] >= rfm['Frequency'].quantile(0.75):
        return "Loyal Customers"
    elif row['Recency'] >= rfm['Recency'].quantile(0.75):
        return "Low Engagement"
    else:
        return "Regular Customers"

rfm['Segment'] = rfm.apply(assign_segment, axis=1)
```

```python
rfm.to_csv("customer_segmentation_final.csv", index=False)
```
